###   A F&B manager wants to determine whether there is any significant difference in the diameter of the cutlet between two units. A randomly selected sample of cutlets was collected from both units and measured? Analyze the data and draw inferences at 5% significance level. Please state the assumptions and tests that you carried out to check validity of the assumptions.​








```python
data=pd.read_csv("Cutlets.csv")
data.head()
```

**Ho (null hypothesis) --->  μ1 = μ2**

**Ha (alternate hypothesis) --->  μ1 != μ2**

```python
t_statistic,p_value=st.ttest_ind(data.iloc[:,0],data.iloc[:,1])
alpha=0.05
if p_value<alpha/2:
    print("Reject null hyothesis")
    print("There is a signigicant difference between sample averages")
else:
    print("Failed to reject null hypothesis")
    print("There is no signigicant difference in the diameter of cutlets between 2 units")
```

# **PROBLEM 2**




### **A hospital wants to determine whether there is any difference in the average Turn Around Time (TAT) of reports of the laboratories on their preferred list. They collected a random sample and recorded TAT for reports of 4 laboratories. TAT is defined as sample collected to report dispatch.**​

   ### **Analyze the data and determine whether there is any difference in average TAT among the different laboratories at 5% significance level.​**

```python
data=pd.read_csv("LabTAT.csv")
data
```

**Ho (null hypothesis) --->  μ1 = μ2 = μ3 = μ4**

**Ha (alternate hypothesis) --->  atleast one is different**

```python
f_statistic,p_value=st.f_oneway(data.iloc[:,0],data.iloc[:,1],data.iloc[:,2],data.iloc[:,3])
alpha=0.05
if p_value<=alpha:
    print("Reject null hypothesis")
    print("Atleast there is a difference in average TAT among the different laboratories")
else:
    print("Reject null hypothesis")
    print("There is no difference in average TAT among the different laboratories")
```

# PROBLEM 3


## **Sales of products in four different regions is tabulated for males and females. Find if male-female buyer rations are similar across regions.​**

```python
data=pd.read_csv("BuyerRatio.csv")
data
```

**Ho (null hypothesis) --->  There is no relation between gender based buyer ratio among regions**

**Ha (alternate hypothesis) --->  There is significant relation between gender based buyer ratio among regions**

```python
chi2_statsistic,p_value,dof,exp=st.chi2_contingency(data.iloc[:,1:])
alpha=0.05
print("chisqaure value :",chi2_statsistic," \np value :",p_value,"\ndof :",dof)
if p_value<alpha:
    print("Reject null hypothesis.\nThere is a significant relation gender based buyers and regions .")
else:
    print("Failed to reject null hypothesis .\nThere is no association between region and gender ratio")
```

# PROBLEM 4


### **TeleCall uses 4 centers around the globe to process customer order forms. They audit a certain %  of the customer order forms. Any error in order form renders it defective and has to be reworked before processing.  The manager wants to check whether the defective %  varies by centre. Please analyze the data at 5% significance level and help the manager draw appropriate inferences​**

```python
data=pd.read_csv("Costomer+OrderForm.csv")
data.head()
```

**Ho (null hypothesis) --->  There is no association between centres and defectives**

**Ha (alternate hypothesis) --->  There is significant relation between centres and defectives**

```python
a=np.array([data[i].value_counts() for i in data.columns]).T
table=pd.DataFrame(a,columns=data.columns,index=["Error Free","Defective"])
table
```

```python
chi2_statsistic,p_value,dof,exp=st.chi2_contingency(table)
alpha=0.05
print("chisqaure value :",chi2_statsistic," \np value :",p_value,"\ndof :",dof)
if p_value<alpha:
    print("Reject null hypothesis.\nThere is a significant relation gender based buyers and regions .")
else:
    print("Failed to reject null hypothesis .\nThere is no association between region and gender ratio")
```
